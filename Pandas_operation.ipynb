## **Indexing and Slicing**

`iloc`



*  iloc stands for 'Integer Location'
*  It is used for index-based selection in a Pandas Dataframe or Series.
* iloc gets rows (or columns) at particular positions in the index (so it only takes integer)


SYNTAX: `df.iloc[row_index, column_index]`



```python
df.iloc[1,]
```

```python
df.head()
```

```python
#4th row's 6th column's value
df.iloc[3,5]
```

```python
#All row's 7th column's value
df.iloc[:,6]
```

```python
#Let us get a subset which consists of first 4 rows and 2 columns
df.iloc[:4,:2]
```

```python
#Let us get s subset which consists of 4th 5th 6th rows and 4th and 5th columns
df.iloc[3:6,3:5]
```

```python
df.head()
```

```python
#Here it will give all rows and columns but sikking 1 row and 1 column
df.iloc[::2, ::2]
```

```python
#Records in reverse order means bottom value at top
df.iloc[::-1,:]
```

```python
#Select from the 3rd last row to the end
df.iloc[-3:,:]
```
